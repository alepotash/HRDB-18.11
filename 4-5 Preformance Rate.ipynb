# KPI Preformance Rate


```python
import pandas as pd
HRDB_cleaned = pd.read_csv('HRDB_cleaned.csv') 
HRDB_cleaned
```

```python
# Same formula concept, Calculates performance rate.

def PreformanceRate(columns, MinEmployees, RowstoPresent):
    rows = []

    for column in columns:
        Categories = HRDB_cleaned[column].dropna().unique()
        
        for cat in Categories:
            mask = HRDB_cleaned[HRDB_cleaned[column] == cat]
            count = len(mask)
            if count < MinEmployees:
                continue
            rate = mask['PerformanceRating'].sum() / count

            rows.append({
                'Category': column,
                'Sub-Category': cat,
                'NumberofEmployees': count,
                'PreformanceRate': rate
            })
    return pd.DataFrame(rows).sort_values(by='PreformanceRate', ascending=False).head(RowstoPresent).round(2)
```

```python
AllCols = HRDB_cleaned.select_dtypes(include=['object']).columns.tolist()
PreformanceRate(AllCols, 10, 50)
```
